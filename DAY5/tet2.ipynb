```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

sns.set_style('darkgrid')
```

```python
#load the simple dataset
df=pd.read_csv('C:\\Users\\User\\Desktop\\Niru\\DAY5\\bok2.csv')
print(df.head())

#visualize the dataset
plt.figure(figsize=(10,6))
plt.scatter(df['Hours'], df['Pass'], c=df['Pass'], cmap='bwr', edgecolor='k')
plt.xlabel('Hours Studied')
plt.ylabel('Result')
plt.title('Hours Studied vs Result')
plt.show()
```

```python
#prepare features and target
x=df[['Hours']]
y=df['Pass']

#initialize and train the model
model=LogisticRegression()
model.fit(x,y)

print("model trained successfully")
```

```python
#visualizing the sigmoid curve (logistic regression curve) 
plt.figure(figsize=(10,6))
plt.scatter(df['Hours'], df['Pass'], c=df['Pass'], cmap='bwr', label="Data Points")

#generate a range of values for the curve
x_test_range=np.linspace(0, 10 , 100).reshape(-1,1)
y_prob=model.predict_proba(x_test_range)[:,1]  #probability of passing

plt.plot(x_test_range, y_prob, color='green', linewidth=3, label='Logistic Regression Curve')
plt.axhline(y=0.5, color='red', linestyle='--', label='Decision Boundary 0.5')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Curve')
plt.legend()
plt.show()
```

```python
try:
    # get user input
    user_hours=float(input("Enter the number of hours studied: "))

    #prepare the input for model
    x_user=np.array([[user_hours]])

    #make prediction
    prediction=model.predict(x_user)[0]
    probability=model.predict_proba(x_user)[0][1]

    #outer result
    if prediction==1:
        print(f"\nResult: Pass")
        print(f"Confidence: {probability:.2f}%")
    else:
        print(f"\nResult: Fail")
        print(f"Confidence: {(1-probability):.2f}%")
except ValueError:
    print("Invalid input. Please enter a numeric value for hours studied.")

```
